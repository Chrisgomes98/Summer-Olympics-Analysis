# Summer Olympics Data Analysis Assignment

### 1. In how many cities Summer Olympics is held so far?

```python
import pandas as pd
df=pd.read_csv("dataset.csv")
print(len(pd.unique(df['City'])))
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
top5Sports=df[df['Medal']=='Gold'].groupby(['Sport']).size().sort_values(ascending=False)[0:5]

df_ = pd.DataFrame(list(zip(list(top5Sports.keys()), list(top5Sports.values))),columns =['Sports', 'GoldsFreq'])
print(df_)
top5Sports.plot.bar()
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
mostMedals=df.groupby('Sport').size().sort_values(ascending=False).head()

df_ = pd.DataFrame(list(zip(list(mostMedals.keys()), list(mostMedals.values))),columns =['Sports', 'MedalCount'])
print(df_)
mostMedals.plot.bar()
```

### 4. Which player has won most number of medals? (Top 5)

```python
mostMedals=df.groupby('Athlete').size().sort_values(ascending=False).head()

df_ = pd.DataFrame(list(zip(list(mostMedals.keys()), list(mostMedals.values))),columns =['Athlete', 'MedalCount'])
print(df_)
mostMedals.plot.bar()
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
top5Sports=df[df['Medal']=='Gold'].groupby(['Athlete']).size().sort_values(ascending=False)[0:5]

df_ = pd.DataFrame(list(zip(list(top5Sports.keys()), list(top5Sports.values))),columns =['Athlete', 'GoldsFreq'])
print(df_)
top5Sports.plot.bar()
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
df_=df[df['Country']=='IND']
df_=df_[df_['Medal']=='Gold']
print("1st year is : ",int(df_.sort_values('Year',ascending=True).head(1)['Year']))
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
top5Sports=df.groupby(['Event']).size().sort_values(ascending=False)[0:5]

df_ = pd.DataFrame(list(zip(list(top5Sports.keys()), list(top5Sports.values))),columns =['Event', 'no. Players'])
print(df_)
top5Sports.plot.bar()
```

```python
df_=df[df['Gender']=='Women']
df_=df_[df_['Medal']=='Gold']
df_=df.groupby('Sport').size().sort_values(ascending=False).head()

print(df_)

df_.plot.bar()
```
